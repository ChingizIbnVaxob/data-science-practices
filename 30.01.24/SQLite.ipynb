```python
df = pd.read_sql('SELECT * FROM customers LIMIT 10', db, index_col = 'CustomerId')
df
```

```python

query = 'SELECT CustomerId, FirstName, LastName, Phone, Country  FROM customers LIMIT 10'
df = pd.read_sql(query, db, index_col = 'CustomerId')
df
```

```python
query = 'SELECT CustomerId, FirstName, LastName, Phone, Country  FROM customers ORDER BY country LIMIT 15'
df = pd.read_sql(query, db, index_col = 'CustomerId')
df
```

```python
query = "SELECT DISTINCT Country FROM customers"
cursor.execute(query)
cursor.fetchall()
```

```python
pd.read_sql(query, db)
```

```python
query = 'SELECT * FROM customers WHERE country="USA" OR country="Canada"'
pd.read_sql(query, db)
```

```python
query = 'SELECT * FROM customers WHERE country="USA" AND State = "CA" ORDER BY Country'
pd.read_sql(query, db)
```
