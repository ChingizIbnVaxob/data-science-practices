```python
import pandas as pd
import numpy as np
```

```python
A = pd.DataFrame({'Transport': ['poezd', 'avtomabil', 'kema'],
                  'Tezligi': ['70', '180', '45']})

A
```

```python
B = pd.DataFrame({'Transport': ['poezd', 'avtomabil', 'tramvay'],
                  'Narxi': ['98000', '10000', '65000']})

B
```

```python
merged = pd.merge(A, B)  # inner join - ichki birlashtirish
```

```python
merged
```

```python
merged_outer = pd.merge(A, B, how='outer')  # outer join - tashqi birlashtirish
merged_outer
```

```python
merged_left = pd.merge(A, B, how='left') # left join - chapdan
merged_left
```

```python
merged_right = pd.merge(A, B, how='right') # left join - o'ngdan
merged_right
```

```python
A = pd.DataFrame({'Transport_turi': ['poezd', 'avtomabil', 'kema'],
                  'Tezligi': ['70', '180', '45']})

A
```

```python
B
```

```python
merg = pd.merge(A, B, left_on = "Transport_turi", right_on = "Transport")
merg
```

# indexlar ustida birlashtirish

```python
merged1 = pd.merge(A, B, left_index=True, right_on = "Transport")
merged1
```
