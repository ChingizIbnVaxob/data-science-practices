```
# Create an array of size (3, 3) with elements from 0 to 9 (9 itself is not an array element) and a step of 1.
```



````python
def nd_array():
    x = np.arange(0, 9, 1).reshape(3,3)
    print(x)
    return x

"""Function check"""
arr = nd_array() # array
assert arr.size == 9, f"there is an error in number of elements" # number of elements
assert type(arr) == np.ndarray, f"array not defined" #numpy array
# elementwise test
test_arr = np.arange(9).reshape(3, 3)
assert arr.shape == (3, 3), f"invalid rows and columns used"
assert True in np.in1d(arr, test_arr), f"there is an interelement error"

def sliced_array_2d():
     """The elements of the array created above are 6 and 7
      section cut function """
     array = nd_array() # create an array using the above function
     print(array[2, 0:2])
     return array[2, 0:2]

"""Function check"""
sl_arr_2d = sliced_array_2d() # sliced array
assert sl_arr_2d.size == 2, f"there is an error in number of elements" # number of elements
assert type(sl_arr_2d) == np.ndarray, f"array not defined" #numpy array
# elementwise test
test_arr = np.array([6, 7])
assert sl_arr_2d.shape == (2,), f"there is an error in the shape of the cropped array"
assert True in np.in1d(sl_arr_2d, test_arr), f"there is an interelement error"
````
