# Heatmap

```python
df2 = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0, parse_dates=[0], usecols=[0, 1] )
df2.head()
```

```python
type(df2.index)
```

```python
df2['month'] = [i.month for i in df2.index]
df2.head()
```

```python
df2['year'] = [i.year for i in df2.index]
df2.head()
```

```python
df3  = df2.groupby(['month', 'year']).max()
df3.head()
```

```python
df3.unstack()
```

```python
df3.unstack(level=0)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df3.unstack(level=0),  cmap='Greens',linewidths=1, square=True)
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                     'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + .5, labels=xticks_labels)
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")
df.head()
```

```python
plt.figure(figsize=(16,6))
sns.set_palette("bright")
sns.barplot(x=df['Month'], y=df['AA'])
plt.title("American Airlines kompaniyasi samolyotlari o'rtacha kechikish vaqti")
plt.ylabel("Minutlar")
plt.show()
```

```python
sns.catplot(data=df, kind="bar", x="Month", y="value", hue="variable")
plt.show()
```

```python

plt.figure(figsize=(14,7))
sns.heatmap(data=df.set_index('Month'), cmap="Accent", annot=True)
plt.xlabel("Airline")
plt.show()
```
