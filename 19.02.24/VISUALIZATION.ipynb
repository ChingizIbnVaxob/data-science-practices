```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(16,6))

plt.plot(x,y, linewidth=2, linestyle='--', color='r', marker='.')
plt.title("sin(x)")
plt.xlabel("pi")
plt.ylabel('sin(x)')
plt.grid(color='k', linestyle="--")
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12,6))

plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color = '#8E44AD',
         alpha=0.7,
         marker='v')

plt.plot(x, cos,
         linewidth = 1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

plt.title('sizn(x) va cos(x) funksiya grafigi')

plt.xlabel('pi')
plt.legend(['sin(x) func', 'cos(x) func'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--')
plt.savefig("grafik1.png")
plt.show()
```
